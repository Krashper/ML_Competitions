# Табличные данные

## Выявление зависимостей

Если зависимая переменная количественная (решаем задачу регрессии)
* признак качественный - ящик с усами для визуализации, стат.тесты (Манна-уитни, Стьюдента, Краскелл-Уоллис)
* признак количественный - точечный график для кадого признака, корреляционная матрица для всех
* однофакторные регрессионные модели (statsmodels)

Если зависимая переменная качественная (решаем задачу классификации)
* признак количественный  - ящик с усами для визуализации, стат.тесты (Манна-уитни, Стьюдента, Краскелл-Уоллис)
* признак качественный - матрица сопряженности, критерий хи-квадрат
* однофакторные классификационные модели (statsmodels)

## Кросс-валидация

* gridsearch (sklearn) - несколько разбиений + тюнинг параметров алгоритма

# Текст

## "Нормализация текста"

* убрать числа и спецсимволы
* перевести в нижний регистр
* разбить текст(документ) на токены
* для каждого токена вернуть нормальную форму (например, с помощью pymorphy2)

## Векторизация

Прошу, не забывайте про настройки min_df, max_df, max_features - они бывают очень важны. Также не забывайте, что есть ещё биграммы, триграммы...

Способы векторизации:
* CountVectorizer
* TF-IDF Vectorizer
* Word2Vec

## Визуализация текстовых данных

Очень классно в случае классификации текстов. Сжимаем признаки до 2 или 3 (можете использовать PCA, например). Рисуем двумерный или трехмерный график. Точки красим в зависимости от класса.

Ну и не забываем про облако слов

## Семантика

* выделение популярных словосочетаний
* тематическое моделирование (например, LDA)

## Классификация текста

После векторизации можете использовать любые известные вам алгоритмы для классификации +

* не забываем про создание модели с помощью Токенизатора Keras
* Добавление LSTM слоя
