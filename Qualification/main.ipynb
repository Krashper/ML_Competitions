## Кластеризация

```python
import sklearn.cluster as cluster
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

standarded_general_data = scaler.fit_transform(general_data) # стандартизируем данные
```

```python
K = range(1,12)
wss = []
for k in K:
    kmeans=cluster.KMeans(n_clusters=k,init="k-means++", random_state=42)
    kmeans=kmeans.fit(standarded_general_data)
    wss_iter = kmeans.inertia_
    wss.append(wss_iter)
```

```python
mycenters = pd.DataFrame({'Clusters' : K, 'WSS' : wss})
mycenters
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x = 'Clusters', y = 'WSS', data = mycenters, marker="+")
plt.show()
```

## Описание числовых данных

```python
from scipy.stats import shapiro, skew
```

```python
def execute_shapiro_test(data):
    result = shapiro(data)
    p_value = result[1]
    
    return p_value

def get_assymetry(data):
    result = skew(data)

    if result > 0.1:
        return "Влево"
    elif result < -0.1:
        return "Вправо"
    return "Не выявлено"

def visual_distribution(data, column):

    fig, ax = plt.subplots(figsize=(12, 5))
    sns.kdeplot(data=data, x=column, ax=ax)
    ax.set_title(f"Распределение для признака: {column}")

    fig.tight_layout()
    plt.show()

    shapiro_p_value = execute_shapiro_test(data[column])
    print(f"По тесту Шапиро нормальность {column}: {shapiro_p_value > 0.05}, p_value: {shapiro_p_value}")

    assymetry = get_assymetry(data[column])
    print(f"Скошенность распределения {column}: {assymetry}")


for col in general_data.columns:
    visual_distribution(general_data, col)
```

4. Временные интервалы поездки и продолжительность

```python
from datetime import datetime
```

```python
def get_trip_interval(line_name: str, start_station: str, end_station: str, start_hour: int):
    try:
        start_date = stations_data[(stations_data["line_name"] == line_name)
                                & (stations_data["station_name"] == start_station)
                                & (stations_data["hour"] == start_hour)][["date", "hour"]]
        if start_date.empty:
            start_date = None
            end_date = None
        else:
            start_date = start_date.iloc[[0]]
            start_index = start_date.index[0]
            filtered_stations_data = stations_data.iloc[start_index+1:]
            

            end_date = filtered_stations_data[(stations_data["line_name"] == line_name)
                                    & (stations_data["date"] == start_date["date"].iloc[0])
                                    & (stations_data["station_name"] == end_station)][["date", "hour"]]
            if end_date.empty:
                end_date = None
            else:
                end_date = end_date.iloc[[0]]
    
        return start_date, end_date
    
    except Exception as e:
        print("Ошибка при получении интервала: ", e)


def get_trip_duration(line_name: str, start_station: str, end_station: str, start_hour: int):
    start_date, end_date = get_trip_interval(line_name, start_station, end_station, start_hour)
    if start_date is None or end_date is None:
        return None, None, None
    trip_duration = end_date["hour"].iloc[0] - start_date["hour"].iloc[0]

    return start_date, end_date, trip_duration
```

## Загрузка данных в базу данных

```python
import psycopg2
from psycopg2 import sql
import csv
```

```python
class DataBase:
    def __init__(self, user_name="postgres", password="", db_name="postgres", host="localhost", port=5432):
        self.user_name = user_name
        self.password = password
        self.db_name = db_name
        self.host = host
        self.port = port
    
    def __get_connect(self):
        conn = psycopg2.connect(user=self.user_name, password=self.password, dbname=self.db_name, host=self.host, port=self.port)
        return conn
    
    def execute_sql_code(self, path):
        conn = self.__get_connect()
        cur = conn.cursor()
        cur.execute(sql.SQL(open(path, "r").read()))
        conn.commit()

    def insert_data(self, filepath: str, table: str):
        try:
            conn = self.__get_connect()
            cur = conn.cursor()
            with open(filepath, 'r', encoding='utf-8') as f:
                reader = csv.reader(f)
                headers = next(reader)
                placeholders = ', '.join(['%s'] * len(headers))
                query = f"INSERT INTO {table} VALUES ({placeholders})"

                total_rows = sum(1 for _ in open(filepath, 'r', encoding='utf-8')) - 1

                for row in tqdm(reader, total=total_rows, desc="Inserting rows"):
                    row = [None if value == '' else value for value in row]
                    cur.execute(query, row)
            conn.commit()
        
        except Exception as e:
            conn.rollback()
            print("Error in during inserting values: ", e)
```
